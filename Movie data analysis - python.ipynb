```python
data.isnull().sum()
# Find The Missing Values
```

```python
data.drop(columns=['homepage'], inplace=True)
```

```python
missing_values=['overview','release_date','runtime','tagline']
for i in missing_values:
    if data[i].dtype=='object':
        data[i].fillna(data[i].mode()[0],inplace=True)
    else:
        data[i].fillna(round(data[i].mean()),inplace=True)
```

# Task 6

```python
# Top 10 movies with the highest revenue
top_10_highest_revenue = data.nlargest(10, 'revenue')

# Top 10 movies with the least budget
top_10_least_budget = data.nsmallest(10, 'budget')

# Display the results
print("Top 10 movies with the highest revenue:")
print(top_10_highest_revenue[['title', 'revenue']])

print("\nTop 10 movies with the least budget:")
print(top_10_least_budget[['title', 'budget']])
```

# The code extracts and displays the top 10 movies with the highest revenue and the top 10 movies with the least budget from a dataset, providing key details about their titles and respective revenue or budget.

# Task 7

```python
# Calculate the Pearson correlation coefficient between popularity and budget
correlation_coefficient = data['popularity'].corr(data['budget'])

# Interpret the correlation coefficient
if correlation_coefficient > 0:
    interpretation = "There is a positive correlation between popularity and budget."
elif correlation_coefficient < 0:
    interpretation = "There is a negative correlation between popularity and budget."
else:
    interpretation = "There is no significant correlation between popularity and budget."

print("Correlation Coefficient:", correlation_coefficient)
print("Interpretation:", interpretation)
```

# The Pearson correlation coefficient between popularity and budget is approximately 0.52, indicating a positive correlation between them.

# Task 8

```python
# Split the 'production_companies' column into a list of production company names
data['production_companies'] = data['production_companies'].str.split('|')

# Create an empty dictionary to store the counts of production companies
production_company_counts = {}

# Iterate through each row and count the appearance of production companies
for companies in data['production_companies']:
    for company in companies:
        if company in production_company_counts:
            production_company_counts[company] += 1
        else:
            production_company_counts[company] = 1

# Convert the dictionary into a DataFrame for easy display
company_counts_data = pd.DataFrame(list(production_company_counts.items()), columns=['Production Company', 'Count'])

# Sort the DataFrame by count in descending order
company_counts_data = company_counts_data.sort_values(by='Count', ascending=False)

print("Production Companies and Their Counts:")
print(company_counts_data)
```

# The code splits the 'production_companies' column into lists of production company names, counts their occurrences, and presents the results in a DataFrame sorted by count in descending order

# Task 10

```python
#  Sort the data in descending order based on revenue
data_sorted_by_revenue = data.sort_values(by='revenue', ascending=False)

# Filter the top 500 movies
top_500_movies = data_sorted_by_revenue.head(500)

# Calculate measures of central tendency
budget_mean = top_500_movies['budget'].mean()
budget_median = top_500_movies['budget'].median()
budget_mode = top_500_movies['budget'].mode().iloc[0]

revenue_mean = top_500_movies['revenue'].mean()
revenue_median = top_500_movies['revenue'].median()
revenue_mode = top_500_movies['revenue'].mode().iloc[0]

runtime_mean = top_500_movies['runtime'].mean()
runtime_median = top_500_movies['runtime'].median()
runtime_mode = top_500_movies['runtime'].mode().iloc[0]
# Display the measures of central tendency
print("Measures of Central Tendency for the Top 500 Movies:")
print("Budget:")
print(f"Mean: {budget_mean}")
print(f"Median: {budget_median}")
print(f"Mode: {budget_mode}")
print("\nRevenue:")
print(f"Mean: {revenue_mean}")
print(f"Median: {revenue_median}")
print(f"Mode: {revenue_mode}")
print("\nRuntime:")
print(f"Mean: {runtime_mean}")
print(f"Median: {runtime_median}")
print(f"Mode: {runtime_mode}")

# Perform outlier analysis using box plots
plt.figure(figsize=(12, 6))

# Box plot for 'budget'
plt.subplot(131)
plt.boxplot(top_500_movies['budget'], vert=False)
plt.title("Budget")

# Box plot for 'revenue'
plt.subplot(132)
plt.boxplot(top_500_movies['revenue'], vert=False)
plt.title("Revenue")

# Box plot for 'runtime'
plt.subplot(133)
plt.boxplot(top_500_movies['runtime'], vert=False)
plt.title("Runtime")

plt.tight_layout()
plt.show()
```

# The box plots show that the top 500 movies have a wide range of budgets, with some outliers on the high end, while their revenue distribution is heavily right-skewed with a few extremely high earners; the runtime distribution is relatively symmetric with no significant outliers
